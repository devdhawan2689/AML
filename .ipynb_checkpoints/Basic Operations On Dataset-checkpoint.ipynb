## Plotting Product Sales Based On Category

```python
productLines = list(data['Product line'].unique())
productLinesCount = []

for category in productLines:
    productLinesCount.append(data['Product line'].value_counts()[category])

productCategory = pd.DataFrame(list(zip(productLines, productLinesCount)), columns=['Product Category', 'Count'])
productCategory
```

```python
productCategory.plot(kind='pie', y='Count', autopct='%1.1f%%', startangle=140, legend=False, ylabel="", title="Category Wise Sales");
```
